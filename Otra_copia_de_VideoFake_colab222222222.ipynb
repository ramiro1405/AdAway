```python
import streamlit as st
import subprocess
import os

st.title("VideoFake Face Swapper")

# Create directories to save uploaded files if they don't exist
UPLOAD_DIR = "uploaded_files"
os.makedirs(UPLOAD_DIR, exist_ok=True)

OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)


def run_face_swapping(source_path, target_path, output_path, video_quality):
    """
    Runs the face swapping script with the given source image and target video.

    Args:
        source_path (str): Path to the source image.
        target_path (str): Path to the target video.
        output_path (str): Path to save the output video.
        video_quality (str): The desired video quality.

    Returns:
        bool: True if the script ran successfully, False otherwise.
    """
    # Change to the directory where the face swapping script is located
    # Ensure we are in the correct directory before running the script
    original_dir = os.getcwd()
    try:
        os.chdir("/content/VideoFake/roop")

        # Define the command to run the face swapping script
        # Using the old command format based on the executed notebook cells
        cmd = f'python run.py --execution-provider cuda --source "{source_path}" -t "{target_path}" -o "{output_path}" --frame-processor face_swapper face_enhancer --output-video-encode libx264 --output-video-quality {video_quality} --keep-fps --many-faces'

        print(f"Running command: {cmd}")

        # Execute the command
        subprocess.run(cmd, shell=True, check=True)
        print("Face swapping script finished successfully.")
        return True
    except subprocess.CalledProcessError as e:
        print(f"Error running face swapping script: {e}")
        return False
    finally:
        # Change back to the original directory
        os.chdir(original_dir)


# Add file upload widgets
source_file = st.file_uploader("Upload Source Image", type=['png', 'jpg', 'jpeg'])
target_file = st.file_uploader("Upload Target Video", type=['mp4', 'avi', 'mov'])

# Add widgets for output parameters
video_quality = st.slider("Video Quality (0-51, lower is better)", 0, 51, 18)


# Add a button to trigger the face swapping process
if st.button("Run Face Swap"):
    if source_file is not None and target_file is not None:
        # Save uploaded files
        source_path = os.path.join(UPLOAD_DIR, source_file.name)
        target_path = os.path.join(UPLOAD_DIR, target_file.name)
        output_path = os.path.join(OUTPUT_DIR, f"swapped_{target_file.name}") # Define output path

        # Basic file type checks (Streamlit's file_uploader type parameter helps, but adding explicit checks is good practice)
        if source_file.type.split('/')[0] != 'image':
            st.error("Invalid source file type. Please upload an image.")
        elif target_file.type.split('/')[0] != 'video':
            st.error("Invalid target file type. Please upload a video.")
        else:
            with open(source_path, "wb") as f:
                f.write(source_file.getbuffer())
            with open(target_path, "wb") as f:
                f.write(target_file.getbuffer())

            # Add placeholders or text elements to display messages
            status_text = st.empty()
            status_text.text("Starting face swap...")

            # Run the face swapping process and check the result
            if run_face_swapping(source_path, target_path, output_path, video_quality):
                status_text.text("Face swap completed successfully! Output video is ready.")
                # Display the output video
                if os.path.exists(output_path):
                     st.video(output_path)
                else:
                     st.error("Output video file not found after processing.")
            else:
                status_text.text("Face swap failed. Please check the Colab notebook output for details.")
    else:
        st.warning("Please upload both a source image and a target video.")
```

```python
%%writefile app.py
import streamlit as st
import subprocess
import os

st.title("VideoFake Face Swapper")

# Create directories to save uploaded files if they don't exist
UPLOAD_DIR = "uploaded_files"
os.makedirs(UPLOAD_DIR, exist_ok=True)

OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)


def run_face_swapping(source_path, target_path, output_path):
    """
    Runs the face swapping script with the given source image and target video.

    Args:
        source_path (str): Path to the source image.
        target_path (str): Path to the target video.
        output_path (str): Path to save the output video.

    Returns:
        bool: True if the script ran successfully, False otherwise.
    """
    # Change to the directory where the face swapping script is located
    os.chdir("/content/VideoFake/roop")

    # Define the command to run the face swapping script
    # Using the old command format based on the executed notebook cells
    cmd = f"python run.py --execution-provider cuda --source {source_path} -t {target_path} -o {output_path} --frame-processor face_swapper face_enhancer --output-video-encode libx264 --output-video-quality 18 --keep-fps --many-faces"

    print(f"Running command: {cmd}")

    try:
        # Execute the command
        subprocess.run(cmd, shell=True, check=True)
        print("Face swapping script finished successfully.")
        return True
    except subprocess.CalledProcessError as e:
        print(f"Error running face swapping script: {e}")
        return False

# Add file upload widgets
source_file = st.file_uploader("Upload Source Image", type=['png', 'jpg', 'jpeg'])
target_file = st.file_uploader("Upload Target Video", type=['mp4', 'avi', 'mov'])

# Add a button to trigger the face swapping process
if st.button("Run Face Swap"):
    if source_file is not None and target_file is not None:
        # Save uploaded files
        source_path = os.path.join(UPLOAD_DIR, source_file.name)
        target_path = os.path.join(UPLOAD_DIR, target_file.name)
        output_path = os.path.join(OUTPUT_DIR, f"swapped_{target_file.name}") # Define output path

        # Basic file type checks (Streamlit's file_uploader type parameter helps, but adding explicit checks is good practice)
        if source_file.type.split('/')[0] != 'image':
            st.error("Invalid source file type. Please upload an image.")
        elif target_file.type.split('/')[0] != 'video':
            st.error("Invalid target file type. Please upload a video.")
        else:
            with open(source_path, "wb") as f:
                f.write(source_file.getbuffer())
            with open(target_path, "wb") as f:
                f.write(target_file.getbuffer())

            # Add placeholders or text elements to display messages
            status_text = st.empty()
            status_text.text("Processing face swap...")

            # Run the face swapping process
            if run_face_swapping(source_path, target_path, output_path):
                status_text.text("Face swap completed successfully!")
                # Display the output video
                if os.path.exists(output_path):
                     st.video(output_path)
                else:
                     st.error("Output video file not found.")
            else:
                status_text.text("Face swap failed.")
    else:
        st.warning("Please upload both a source image and a target video.")
```
